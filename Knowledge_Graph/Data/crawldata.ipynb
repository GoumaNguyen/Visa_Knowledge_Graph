```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import os
import json
import time

# Đọc file CSV
csv_file = "./repealed_visa_links.csv"  # Đường dẫn tới file CSV của bạn
df = pd.read_csv(csv_file)

# Thiết lập thư mục lưu JSON
save_folder = "./crawled/"
if not os.path.exists(save_folder):
    os.makedirs(save_folder)

# Thiết lập ChromeDriver (sử dụng driver có sẵn)
service = Service(ChromeDriverManager().install())  
driver = webdriver.Chrome(service=service)

# Hàm crawl nội dung từ trang web
def crawl_content(link_crawl):
    # Mở trang web
    driver.get(link_crawl)
    
    # Đợi trang tải
    time.sleep(3)  # Điều chỉnh thời gian nếu cần
    
    content = ""
    
    try:
        # Lấy nội dung từ class="apply-direction"
        apply_direction_tags = driver.find_elements(By.CLASS_NAME, "apply-direction")
        if apply_direction_tags:
            content += "\n".join([tag.text for tag in apply_direction_tags])
        
        # Lấy nội dung từ các thẻ p có id="ctl00_PlaceHolderMain_ctl03__ControlWrapper_RichHtmlField"
        rich_html_field = driver.find_elements(By.CSS_SELECTOR, "#ctl00_PlaceHolderMain_ctl03__ControlWrapper_RichHtmlField p")
        if rich_html_field:
            content += "\n".join([tag.text for tag in rich_html_field])
        
        # Lấy nội dung từ các thẻ có class=".edit-text"
        edit_text_tags = driver.find_elements(By.CSS_SELECTOR, ".edit-text")
        if edit_text_tags:
            content += "\n".join([tag.text for tag in edit_text_tags])
        
        if not content:
            content = "No content found"
    
    except Exception as e:
        content = f"Error during crawl: {e}"
    
    return content

# Lặp qua từng dòng trong file CSV
for index, row in df.iterrows():
    visaID = row['visaID']
    visaType = row['visaType']
    visaName = row['visaName']
    link_crawl = row['link_crawl']
    
    # Crawl dữ liệu từ link_crawl
    content = crawl_content(link_crawl)
    
    # Chuẩn bị dữ liệu JSON
    data = {
        "visaID": visaID,
        "visaType": visaType,
        "visaName": visaName,
        "link_crawl": link_crawl,
        "content": content
    }
    
    # Tạo tên file lưu với visaID
    json_file = os.path.join(save_folder, f"{visaID}.json")
    
    # Lưu dữ liệu JSON vào file
    with open(json_file, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    
    print(f"Data saved to {json_file}")
    

# Đóng trình duyệt sau khi hoàn thành
driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import os
import json
import time

# Đọc file CSV
csv_file = "./Visa_data.csv"  # Đường dẫn tới file CSV của bạn
df = pd.read_csv(csv_file)

# Thiết lập thư mục lưu JSON
save_folder = "./crawled_v1/"
if not os.path.exists(save_folder):
    os.makedirs(save_folder)

# Thiết lập ChromeDriver (sử dụng driver có sẵn)
service = Service(ChromeDriverManager().install())  
driver = webdriver.Chrome(service=service)

def crawl_content(link_crawl):
        # Tạo đường dẫn crawl
    link_crawl = link_crawl + "#About"
    # Mở trang web
    driver.get(link_crawl)
    
    # Đợi trang tải
    time.sleep(3)  # Điều chỉnh thời gian nếu cần
    
    content = ""
    span_count = 1  # Biến đếm cho thẻ span để đánh chỉ mục
    
    try:
        # Lấy nội dung từ class="apply-direction"
        apply_direction_tags = driver.find_elements(By.CLASS_NAME, "apply-direction")
        if apply_direction_tags:
            content += "\n".join([tag.text for tag in apply_direction_tags])
        
        # Lấy nội dung từ các thẻ p có id="ctl00_PlaceHolderMain_ctl03__ControlWrapper_RichHtmlField"
        rich_html_field = driver.find_elements(By.CSS_SELECTOR, "#ctl00_PlaceHolderMain_ctl03__ControlWrapper_RichHtmlField p")
        if rich_html_field:
            content += "\n".join([tag.text for tag in rich_html_field])
        
        # Lấy các thẻ có class=".edit-text"
        edit_text_tags = driver.find_elements(By.CSS_SELECTOR, ".edit-text")
        for tag in edit_text_tags:
            # Xử lý các thẻ span có nội dung trong class edit-text
            if tag.tag_name == "span" and tag.text.strip():
                # Đánh số thứ tự cho các thẻ span có nội dung
                content += f"\n{span_count}. {tag.text.strip()}"
                span_count += 1
            
            # Lấy nội dung của các thẻ p trong div.edit-text
            p_tags = tag.find_elements(By.TAG_NAME, "p")
            for p_tag in p_tags:
                text = driver.execute_script("return arguments[0].innerText;", p_tag).strip()
                if text:
                    content += f"\n{text}"
        
        if not content:
            content = "No content found"
    
    except Exception as e:
        content = f"Error during crawl: {e}"
    
    return content


# Lặp qua từng dòng trong file CSV
for index, row in df.iterrows():
    visaID = row['visaID']
    visaType = row['visaType']
    visaName = row['visaName']
    link_crawl = row['link_crawl']
    
    # Crawl dữ liệu từ link_crawl
    content = crawl_content(link_crawl)
    
    # Chuẩn bị dữ liệu JSON
    data = {
        "visaID": visaID,
        "visaType": visaType,
        "visaName": visaName,
        "link_crawl": link_crawl,
        "content": content
    }
    
    # Tạo tên file lưu với visaID
    json_file = os.path.join(save_folder, f"{visaID}.json")
    
    # Lưu dữ liệu JSON vào file
    with open(json_file, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    
    print(f"Data saved to {json_file}")
    

# Đóng trình duyệt sau khi hoàn thành
driver.quit()
```

```python
import os
import re

def extract_numbered_lines(folder_path):
    numbered_lines = set()  # Sử dụng set để loại bỏ các dòng trùng lặp

    # Duyệt qua các file trong thư mục
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):  # Chỉ xử lý các file .txt
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                for line in file:
                    # Kiểm tra nếu dòng bắt đầu bằng định dạng "number + '.'"
                    match = re.match(r'^\d+\.\s', line)
                    if match:
                        # Xóa số, dấu chấm và khoảng trắng phía sau
                        content = line[match.end():].strip()
                        numbered_lines.add(content)  # Thêm nội dung đã loại bỏ số vào set

    return numbered_lines

# Thư mục chứa các file .txt
folder_path = './content only ver4/'

# Lấy các dòng có định dạng yêu cầu và in ra
numbered_lines = extract_numbered_lines(folder_path)
for line in sorted(numbered_lines):  # Sắp xếp để dễ nhìn hơn
    print(line)
```

```python
import os
import re

def process_files_and_copy(src_folder, dest_folder):
    # Tạo thư mục đích nếu chưa tồn tại
    os.makedirs(dest_folder, exist_ok=True)
    
    # Duyệt qua từng file trong thư mục nguồn
    for filename in os.listdir(src_folder):
        if filename.endswith(".txt"):  # Chỉ xử lý các file .txt
            src_file_path = os.path.join(src_folder, filename)
            dest_file_path = os.path.join(dest_folder, filename)
            
            # Đọc file và thực hiện xử lý
            with open(src_file_path, 'r', encoding='utf-8') as file:
                lines = file.readlines()

            if lines:
                # Xử lý dòng đầu tiên
                first_line = lines[0].strip()
                print(f"Original First Line: {first_line}")
                
                # Loại bỏ dấu ngoặc ( ) nhưng giữ lại nội dung bên trong
                cleaned_line = re.sub(r'[()]', '', first_line)
                print(f"Cleaned Line: {cleaned_line}")
                
                # Loại bỏ index "1." nếu có
                cleaned_line = re.sub(r'^\d+\.\s*', '', cleaned_line)
                print(f"Line without Index: {cleaned_line}")
                
                # Cập nhật regex để nhận diện cả trường hợp có ngoặc xung quanh "Subclass XYZ"
                match = re.match(r'^\(?Subclass (\d+)\)?$', cleaned_line.strip())
                
                if match:
                    subclass_text = match.group(1)
                    subclass_info = f"Subclass {subclass_text}"
                    print(f"Matched Subclass: {subclass_info}")
                    
                    # Đưa subclass_info xuống cuối dòng thứ hai nếu có
                    if len(lines) > 1:
                        lines[1] = lines[1].strip() + f" ({subclass_info})\n"
                        print(f"Updated Second Line: {lines[1]}")
                    else:
                        lines.append(f"({subclass_info})\n")

            # Ghi lại nội dung đã chỉnh sửa vào file đích trong thư mục ver5
            with open(dest_file_path, 'w', encoding='utf-8') as dest_file:
                dest_file.writelines(lines)

# Thư mục nguồn và thư mục đích
src_folder = './content only ver4/'
dest_folder = './content only ver5/'
process_files_and_copy(src_folder, dest_folder)
```

```python
import os
import json
import re

# Đường dẫn tới thư mục chứa các file JSON
folder_path = "./crawled_v1/"
folder_save_path = "./content only ver6/"

# Tạo thư mục đích nếu chưa tồn tại
os.makedirs(folder_save_path, exist_ok=True)

# Duyệt qua từng file trong thư mục
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)
        
        # Đọc nội dung file JSON
        with open(file_path, "r", encoding="utf-8") as file:
            data = json.load(file)
            visa_id = data.get("visaID")
            content = data.get("content", "")

            # Loại bỏ các dòng trống liên tiếp trong content
            cleaned_content = re.sub(r'\n\s*\n+', '\n\n', content.strip())
            
            # Tạo đường dẫn file .txt mới trong thư mục đích với tên file theo visa_id
            output_file_path = os.path.join(folder_save_path, f"{visa_id}.txt")
            
            # Ghi content đã xử lý vào file .txt riêng cho từng visa_id
            with open(output_file_path, "w", encoding="utf-8") as output_file:
                output_file.write(cleaned_content)

print("Đã loại bỏ dòng trống liên tiếp và lưu các file .txt riêng cho từng visaID.")
```
