```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import json
import os
import time
from webdriver_manager.chrome import ChromeDriverManager  # Thêm dòng này để tự động cài đặt ChromeDriver

# Đọc file CSV
csv_file = "Visa_data.csv"  # Đường dẫn đến file CSV của bạn
df = pd.read_csv(csv_file)

# Thiết lập ChromeDriver (sử dụng driver có sẵn)
service = Service(ChromeDriverManager().install())  # Dùng ChromeDriverManager để tự động cài đặt
driver = webdriver.Chrome(service=service)

# Tạo thư mục lưu nếu chưa có
save_path = "./crawled/"
if not os.path.exists(save_path):
    os.makedirs(save_path)

# Duyệt qua từng dòng trong DataFrame
for index, row in df.iterrows():
    visaID = row['visaID']
    visaType = row['visaType']
    visaName = row['visaName']
    link_crawl = row['link_crawl']

    # Tạo đường dẫn crawl
    crawl_path = link_crawl + "#About"

    try:
        # Mở trang web
        driver.get(crawl_path)
        
        # Đợi trang tải
        time.sleep(3)  # Điều chỉnh thời gian nếu cần
        
        # Lấy toàn bộ nội dung từ các thẻ có class '.edit-text'
        content_tags = driver.find_elements(By.CSS_SELECTOR, '.edit-text')
        
        # Nếu tìm thấy thẻ '.edit-text', nối tất cả nội dung
        if content_tags:
            content_text = "\n".join([tag.text for tag in content_tags])
        else:
            content_text = "No content found"

    except Exception as e:
        # Nếu có lỗi khi crawl, ghi lại lỗi
        content_text = f"Error occurred: {str(e)}"

    # Chuẩn bị dữ liệu JSON
    data = {
        "visaID": visaID,
        "visaType": visaType,
        "visaName": visaName,
        "link_crawl": link_crawl,
        "content": content_text
    }

    # Tạo tên file lưu với visaID
    json_file = os.path.join(save_path, f"{visaID}.json")

    # Lưu dữ liệu JSON vào file
    with open(json_file, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

    print(f"Data saved to {json_file}")

# Đóng trình duyệt sau khi đã crawl toàn bộ dữ liệu
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Initialize the Chrome driver (or another driver if preferred)
driver = webdriver.Chrome()  # Ensure the correct path to the driver

# Open the page
url = "https://immi.homeaffairs.gov.au/visas/getting-a-visa/visa-listing"
driver.get(url)

# Allow some time for the page to load
time.sleep(5)

# Find all elements with the class 'edit-text'
text_elements = driver.find_elements(By.CLASS_NAME, 'edit-text')

# Write content to a text file
with open('visa_listing.txt', 'w', encoding='utf-8') as f:
    for elem in text_elements:
        f.write(elem.text.strip() + '\n')

print("Content saved to visa_listing.txt")

# Close the browser
driver.quit()
```
